## psuedo algorithm

1. start with (2, 3) as the primary list, and n = 4
2. add 2 to n, until n cannot be generated as a sum of any of the terms in the list
3. create a new list by subtracting each term in the primary list from n
4. take the smallest prime from the generated list, and add it to the primary list
5. repeat step 2

```python
def is_sum_of_two_elements(n, arr):
    for a in arr[::-1]:
        for b in arr[::-1]:
            if (a + b == n):
                return True
    return False
```

```python
def step_2(n, primary_list):
    while (is_sum_of_two_elements(n, primary_list)):
        n += 2
    return n
```

```python
def step_3(n, primary_list):
    acc = []
    for p in primary_list:
        acc.append(n - p)
    return acc
```

```python
def is_prime(n):
    for a in range(2, int(n/2)):
        if (n % a == 0):
            return False
    return True
```

```python
def smallest_prime(arr):
    primes = []
    for n in arr:
        if (is_prime(n)):
            primes.append(n)
    return min(primes)
```

```python
primary_list = [2, 3]
ranges = []
n = 4
old_n = n
count = 0
```

```python
while (count < 100):
    count+=1
    n = step_2(n, primary_list)
    print(f"got to {n}, with a range of {n - old_n}", end='')
    ranges.append(n - old_n)
    old_n = n
    new = smallest_prime(step_3(n, primary_list))
    print(f" appending {new}")
    primary_list.append(new)
#     print(primary_list)
```
